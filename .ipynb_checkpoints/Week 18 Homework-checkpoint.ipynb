### Introduction to Neural Networks

## 1. What is a neural network? What are the general steps required to build a neural network?

A neural network is a complex predictive model. The models we have worked with up to this point have considered how input features interact with the output feature. Neural network add layers of understanding that learn the interactions between input features, themselves, and how those combinations of interactions interact with the target variable to make better real-world predictions. Generally, to build a neural network, you separate your input features from your output. Then you select and instantiate the model. You next add hidden layers (or several) of nodes, model constructors and layer constructors. Then you compile your model with the help of an optimizer and a loss function. Next, you create activation functions to apply backpropagation and gradient descent. This will bounce the predictions forward and backward from input layer thru hidden layers to output layer over several iterations, or epochs, where the neural network is adding to it's abilities to capture interactions between features and adjusting their associated weights with each pass. Hopefully this culminates in the lowest possible error, but sometimes a stop condition is used to cap the processing, instead.

## 2. Generally, how do you check the performance of a neural network? Why? 

In general, a loss function is used. This is the difference between the actual values and what the model has predicted. This is what the neural network uses to learn and adjust weights for features and interactions. Unless stopped, the model will continue to run until it has reached the lowest loss value possible with it's parameters. If your loss isn't getting as low as you'd expect, you know you need to tune the neural network itself and probably not your data. Maybe your step is too large in your gradient descent and you are missing the bottom of the loss curve, maybe your learning value is too high or too low. Regardless, loss is the metric that is used to evaluate, understand, and tune neural networks.

## 5. Create a neural network using pytorch to predict the same result as question 3. 

### Pytorch NN

### Cleaned data:

```python
# using method from https://machinelearningmastery.com/pytorch-tutorial-develop-deep-learning-models/
# This way uses torch packages instead of the pandas or sklearn functions that I've been using:

# Make the dataset into a Class with a torch function (just trying this out for practice)::
class AbaloneCleaned(Dataset):
    # load the dataset with a function:  (just trying this out for practice):
    def __init__(self, path):
        # load the csv file as a dataframe
        abalone_cleaned = pd.read_csv(path, header=None)
        # store the inputs and outputs
        self.X = abalone_cleaned.values[:, :-1].astype('float32')
        self.y = abalone_cleaned.values[:, -1].astype('float32')
        # ensure target has the right shape
        self.y = self.y.reshape((len(self.y), 1))
 
    # setting number of rows in the dataset:
    def __len__(self):
        return len(self.X)
 
     # function to get a row at a particular index:
    def __getitem__(self, idx):
        return [self.X[idx], self.y[idx]]
 
    # function to get indexes for the train and test rows from random_split:
    def get_splits(self, n_test=0.33):
        # determine sizes, like, with math.
        test_size = round(n_test * len(self.X))
        train_size = len(self.X) - test_size
        return random_split(self, [train_size, test_size])
 # So, train_test_split is faster and easier, but I'll give this a try and see if I like it.
```

```python
# defining the model:
class MLP(Module):
    # choose elements for the model:
    def __init__(self, n_inputs):
        super(MLP, self).__init__()
        # design my first hidden layer (*squeeee!*)
        self.hidden1 = Linear(n_inputs, 10)
        # using the xavier_uniform_ torch function to set my tensors (totally had to research this...)
        xavier_uniform_(self.hidden1.weight)
        self.act1 = Sigmoid()# <-- I'm excited to use this as my activation function, because in 
        # other iterations of my model, I kept running into the "dying neuron problem" and one of the
        # advantages of this function is that it keeps "close to 0" instead of 0, which kills the node.
        # SECOND hidden layer:
        self.hidden2 = Linear(10, 8)
        xavier_uniform_(self.hidden2.weight)
        self.act2 = Sigmoid()
        # THIRD hidden layer and OUTPUT:
        self.hidden3 = Linear(8, 1)
        xavier_uniform_(self.hidden3.weight)
 
    # Forward propogation input (oooh, this is exciting):
    def forward(self, X):
        # input to first hidden layer
        X = self.hidden1(X)
        X = self.act1(X)
         # second hidden layer
        X = self.hidden2(X)
        X = self.act2(X)
        # third hidden layer and output
        X = self.hidden3(X)
        return X
```

```python
# Preparing the dataset:
def prepare_data(path):
    # load the dataset
    dataset = AbaloneCleaned(path)
    # get split:
    train, test = dataset.get_splits()
    # use data loaders:
    train_dl = DataLoader(train, batch_size=32, shuffle=True)
    test_dl = DataLoader(test, batch_size=1400, shuffle=False)# I think I want test_dl to have
    # a length of 1, so my batch size should be lenght of my test set or 4177#0.33=1378.41 ~ 1400.
    return train_dl, test_dl
 
# train the model
def train_model(train_dl, model):
    # choosing my optimization:
    criterion = MSELoss()
    optimizer = Adam(model.parameters(), lr=0.001)
    # enumerate epochs
    for epoch in range(800):
        # enumerate mini batches
        torch_loss_values = list()
        for i, (inputs, targets) in enumerate(train_dl):
            # reset the gradients
            optimizer.zero_grad()
            # get my output
            y_torch_pred = model(inputs)
            # find loss
            loss = criterion(y_torch_pred, targets)
            # store final value:
            torch_loss_values.append(loss)
            # fire up backward propogation:
            loss.backward()
            # update model weights
            optimizer.step()
    torch_final_loss = torch_loss_values[-1]
    return final_loss
```

```python
 # EVALUATE! :D
def evaluate_model(test_dl, model):
    predictions, actuals = list(), list()
    for i, (inputs, targets) in enumerate(test_dl):
        # run the test set!
        y_torch_pred = model(inputs)
        # generate an array
        y_torch_pred = y_torch_pred.detach().numpy()# < Never seen detach() before...
        actual = targets.numpy()
        actual = actual.reshape((len(actual), 1))
        # store in the lists:
        predictions.append(y_torch_pred)
        actuals.append(actual)
    predictions, actuals = vstack(predictions), vstack(actuals)# < Never seen this before, either.
    # Get metrics:
    torch_rmse = np.sqrt(mean_squared_error(actuals, predictions))
    return torch_rmse
```

```python
# CALLING ALL FUNCTIONS! ALL ABOARD! LEAVING THE STATION!:

path = 'abalone_df.csv'
train_dl, test_dl = prepare_data(path)

# define the network
model = MLP(8)

# train the model
train_model(train_dl, model)

# evaluate the model
torch_rmse = evaluate_model(test_dl, model)
torch_final_loss = train_model(train_dl, model)
print('RMSE: %.4f, Loss: %.5f' % (torch_rmse, final_loss))

# Audible cell-completion notification:
Audio('Bicycle-bell.mp3', autoplay=True)
```

```python
# Make the dataset: 
class AbalonePP(Dataset):
    # load the dataset with a function: 
    def __init__(self, pp_path):
        # load the csv file as a dataframe
        abalone_pp = pd.read_csv(pp_path, header=None)
        # store the inputs and outputs
        self.X = abalone_pp.values[:, :-1].astype('float32')
        self.y = abalone_pp.values[:, -1].astype('float32')
        # ensure target has the right shape
        self.y = self.y.reshape((len(self.y), 1))
 
    # setting number of rows in the dataset:
    def __len__(self):
        return len(self.X)
 
     # function to get a row at a particular index:
    def __getitem__(self, idx):
        return [self.X[idx], self.y[idx]]
 
    # function to get indexes for the train and test rows from random_split:
    def get_splits(self, n_test=0.33):
        # determine sizes, like, with math.
        test_size = round(n_test * len(self.X))
        train_size = len(self.X) - test_size
        return random_split(self, [train_size, test_size])
```

```python
# defining the model:
class pp_MLP(Module):
    # choose elements for the model:
    def __init__(self, n_inputs):
        super(MLP, self).__init__()
        # design my first hidden layer:
        self.hidden1 = Linear(n_inputs, 10)
        xavier_uniform_(self.hidden1.weight)
        self.act1 = Sigmoid()
        # SECOND hidden layer:
        self.hidden2 = Linear(10, 8)
        xavier_uniform_(self.hidden2.weight)
        self.act2 = Sigmoid()
        # THIRD hidden layer and OUTPUT:
        self.hidden3 = Linear(8, 1)
        xavier_uniform_(self.hidden3.weight)
 
    # Forward propogation input:
    def forward(self, X):
        # input to first hidden layer
        X = self.hidden1(X)
        X = self.act1(X)
         # second hidden layer
        X = self.hidden2(X)
        X = self.act2(X)
        # third hidden layer and output
        X = self.hidden3(X)
        return X
```

```python
# Preparing the dataset:
def prepare_pp_data(pp_path):
    # load the dataset
    dataset = AbalonePP(pp_path)
    # get split:
    train, test = dataset.get_splits()
    # use data loaders:
    train_pp_dl = DataLoader(train, batch_size=32, shuffle=True)
    test_pp_dl = DataLoader(test, batch_size=1400, shuffle=False)
    return train_dl, test_dl
 
# train the model
def train_pp_model(train_pp_dl, pp_model):
    # choosing my optimization:
    criterion = MSELoss()
    optimizer = Adam(pp_model.parameters(), lr=0.01)
    # enumerate epochs
    for epoch in range(800):
        # enumerate mini batches
        torch_loss_values = list()
        for i, (inputs, targets) in enumerate(train_pp_dl):
            # reset the gradients
            optimizer.zero_grad()
            # get my output
            y_torch_pred = pp_model(inputs)
            # find loss
            loss = criterion(y_torch_pred, targets)
            # store final value:
            torch_loss_values.append(loss)
            # fire up backward propogation:
            loss.backward()
            # update model weights
            optimizer.step()
    pp_torch_final_loss = torch_loss_values[-1]
    return pp_torch_final_loss
```

```python
 # EVALUATE! :D
def evaluate_pp_model(test_pp_dl, pp_model):
    predictions, actuals = list(), list()
    for i, (inputs, targets) in enumerate(test_pp_dl):
        # run the test set!
        y_torch_pred = pp_model(inputs)
        # generate an array
        y_torch_pred = y_torch_pred.detach().numpy()# < Never seen detach() before...
        actual = targets.numpy()
        actual = actual.reshape((len(actual), 1))
        # store in the lists:
        predictions.append(y_torch_pred)
        actuals.append(actual)
    predictions, actuals = vstack(predictions), vstack(actuals)# < Never seen this before, either.
    # Get metrics:
    pp_torch_rmse = np.sqrt(mean_squared_error(actuals, predictions))
    return pp_torch_rmse
```

```python
# CALLING ALL FUNCTIONS! ALL ABOARD! LEAVING THE STATION!:

pp_path = 'abalone.csv'
train_pp_dl, test_pp_dl = prepare_pp_data(pp_path)

# define the network
pp_model = MLP(5)

# train the model
train_pp_model(train_pp_dl,pp_model)

# evaluate the model
pp_torch_rmse = evaluate_pp_model(test_pp_dl, pp_model)
pp_torch_final_loss = train_pp_model(train_pp_dl, pp_model)
print('RMSE: %.4f, Loss: %.5f' % (pp_torch_rmse, pp_torch_final_loss))

# Audible cell-completion notification:
Audio('Bicycle-bell.mp3', autoplay=True)
```
